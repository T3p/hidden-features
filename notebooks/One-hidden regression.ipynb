```python
from hiddenfeat.bandits.finite_linear_bandits import make_random_hflb
from hiddenfeat.algos.one_feat_regression import solve, solve_auto
import matplotlib.pyplot as plt
from hiddenfeat.utils import seed_all
```

```python
iterations = 1000
seed_all(0)
bandit = make_random_hflb(10, 10, dim=10, noise=0.1, hidden=1)
```

### Manual

```python
param, hf, param_errors, pred_errors = solve(bandit, d=1, lambda1=0.1, lambda2=0.1, samples=iterations)
```

```python
plt.plot(range(iterations), param_errors)
```

```python
plt.plot(range(iterations), pred_errors)
```

### Auto

```python
param, hf, param_errors, pred_errors = solve_auto(bandit, d=1, lambda1=0.01, lambda2=0.01, samples=iterations)
```

```python
plt.plot(range(iterations), param_errors)
```

```python
plt.plot(range(iterations), pred_errors)
```
